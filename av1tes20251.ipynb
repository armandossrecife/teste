# Faz as comparações

```python
# prompt: Crie uma função que dado o dataframe df_alunos_tes compare se NOME aparece na lista1 de nomes, se fim atualiza a coluna l1 para 1 se não atualiza a coluna l1 para 0

def comparar_nomes_lista(df_alunos, lista_nomes, avaliacao):
  if avaliacao=='l1':
    df_alunos['l1'] = df_alunos['NOME'].apply(lambda x: 1 if x in lista_nomes else 0)
  if avaliacao=='l2':
    df_alunos['l2'] = df_alunos['NOME'].apply(lambda x: 1 if x in lista_nomes else 0)
  if avaliacao=='l3':
    df_alunos['l3'] = df_alunos['NOME'].apply(lambda x: 1 if x in lista_nomes else 0)
  if avaliacao=='l4':
    df_alunos['l4'] = df_alunos['NOME'].apply(lambda x: 1 if x in lista_nomes else 0)

  return df_alunos
```

```python
df_alunos_tes = comparar_nomes_lista(df_alunos_tes, lista1, 'l1')
df_alunos_tes = comparar_nomes_lista(df_alunos_tes, lista2, 'l2')
df_alunos_tes = comparar_nomes_lista(df_alunos_tes, lista3, 'l3')
df_alunos_tes = comparar_nomes_lista(df_alunos_tes, lista4, 'l4')
```

```python
# prompt: Como gerar uma nova coluna primeiro_nome no dataframe df_alunos_tes usando a coluna NOME

df_alunos_tes['primeiro_nome'] = df_alunos_tes['NOME'].apply(lambda x: x.split(' ')[0])
df_alunos_tes
```

```python
def testar_nome_nas_listas(nome):
  t1, t2, t3, t4 = '', '', '', ''
  for i in lista1:
    if nome in i:
      t1 = 'l1 OK!'
  for i in lista2:
    if nome in i:
      t2 = 'l2 OK!'
  for i in lista3:
    if nome in i:
      t3 = 'l3 OK!'
  for i in lista4:
    if nome in i:
      t4 = 'l4 OK!'
  resultado = t1 + ',' + t2 + ',' + t3 + ',' + t4
  return resultado
```

```python
df_alunos_tes['listas'] = df_alunos_tes['primeiro_nome'].apply(lambda x: testar_nome_nas_listas(x))
df_alunos_tes
```
