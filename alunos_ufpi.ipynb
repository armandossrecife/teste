```python
import pandas as pd

alunos_computacao = pd.read_html('https://www.sigaa.ufpi.br/sigaa/public/curso/alunos.jsf?lc=pt_BR&id=74268')
alunos_computacao[0]
```

```python
alunos_computacao[0].info()
```

```python
df_alunos_computacao_2023_2_ativos = alunos_computacao[0][['Matrícula', 'Aluno']]
df_alunos_computacao_2023_2_ativos
```

```python
df_alunos_computacao_2023_2_ativos.drop(df_alunos_computacao_2023_2_ativos.tail(1).index,inplace=True) # drop last n rows
```

```python
df_alunos_computacao_2023_2_ativos
```

```python
df_alunos_computacao_2023_2_ativos.to_csv('alunos_computacao_ativos_2023.csv', index=False)
```
