# Faz a análise...

```python
def carrega_alunos(arquivo):
  l1 = list()
  try:
    f_l1 = open(arquivo, 'r')
    for linha in f_l1:
      l1.append(linha.strip())
    f_l1.close() 
  except Exception as ex: 
    print(f'Erro ao carregar arquivo {arquivo}: {str(ex)}')
  return l1
```

```python
l_alunos_lp = []
f_alunos_lp = open('alunoslp.txt')
for linha in f_alunos_lp:
  l_alunos_lp.append(linha.strip())
f_alunos_lp.close()

f_alunos_csv = open('alunos_lp.csv', 'w')
f_alunos_csv.write('id, matricula, nome')
f_alunos_csv.write('\n')
l_nome_completo = list()
for each in l_alunos_lp:
  each = each.split(' ')
  id = each[0]
  matricula = each[1]
  nome = each[2:]
  nome_completo = " ".join(str(e) for e in nome)
  l_nome_completo.append(nome_completo)
  nova_linha = id + ',' + matricula + ',' + nome_completo + '\n'
  f_alunos_csv.write(nova_linha)
f_alunos_csv.close()

lista1 = carrega_alunos(arquivo='l1.txt')
lista2 = carrega_alunos(arquivo='l2.txt')
lista3 = carrega_alunos(arquivo='l3.txt')
lista4 = carrega_alunos(arquivo='l4.txt')

print(f'Total de alunos: {len(l_nome_completo)}')
print(f'Resolução - l1: {len(lista1)}, l2: {len(lista2)}, l3: {len(lista3)}, l4: {len(lista4)}')
```

```python
fez_l1, fez_l2, fez_l3, fez_l4 = ([], [], [], [])
for nome in l_nome_completo:
  if nome in lista1:  
    fez_l1.append(1)
  else:
    fez_l1.append(0)
  if nome in lista2:  
    fez_l2.append(1)
  else:
    fez_l2.append(0)
  if nome in lista3:  
    fez_l3.append(1)
  else:
    fez_l3.append(0)
  if nome in lista4:  
    fez_l4.append(1)
  else:
    fez_l4.append(0)
```

```python
dic_alunos_lp = {
    'nome':l_nome_completo,
    'fez_l1':fez_l1, 
    'fez_l2':fez_l2, 
    'fez_l3':fez_l3, 
    'fez_l4':fez_l4, 
}
```
