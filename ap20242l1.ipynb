# 1. Ler os dados das listas

```python
with open('lista1.txt', mode='r') as f1:
    lista1 = f1.read().splitlines()
with open('lista2.txt', mode='r') as f2:
    lista2 = f2.read().splitlines()
with open('lista3.txt', mode='r') as f3:
    lista3 = f3.read().splitlines()
with open('lista4.txt', mode='r') as f4:
    lista4 = f4.read().splitlines()
```

```python
def limpa_nomes(lista):
  l1 = list()
  for each in lista:
    each = each.strip()
    l1.append(each)
  return l1
```

```python
l1 = limpa_nomes(lista1)
l2 = limpa_nomes(lista2)
l3 = limpa_nomes(lista3)
l4 = limpa_nomes(lista4)
```

# 2. Ler a lista de presen√ßa

```python
import pandas

df_alunos = pandas.read_csv('alunosap.csv')
df_alunos
```

```python
# prompt: convert df_alunos to disctionary where key is matricula and value is nome.

alunos_dict = dict(zip(df_alunos['matricula'], df_alunos['nome']))
alunos_dict
```

#3. Cria as listas de respostas

```python
l1_respostas = list()
l2_respostas = list()
l3_respostas = list()
l4_respostas = list()
```

```python
for k,v in alunos_dict.items():
  print(k,v)
  if v in l1:
    l1_respostas.append(1)
  else:
    l1_respostas.append(0)
  if v in l2:
    l2_respostas.append(1)
  else:
    l2_respostas.append(0)
  if v in l3:
    l3_respostas.append(1)
  else:
    l3_respostas.append(0)
  if v in l4:
    l4_respostas.append(1)
  else:
    l4_respostas.append(0)
```

```python
len(l1_respostas), len(l2_respostas), len(l3_respostas), len(l4_respostas)
```

```python
# prompt: how to put l1_respostas as a new column in df_alunos?

df_alunos['l1_respostas'] = l1_respostas
df_alunos['l2_respostas'] = l2_respostas
df_alunos['l3_respostas'] = l3_respostas
df_alunos['l4_respostas'] = l4_respostas
df_alunos
```

```python
df_alunos['soma'] = df_alunos['l1_respostas'] + df_alunos['l2_respostas'] + df_alunos['l3_respostas'] + df_alunos['l4_respostas']
df_alunos
```
