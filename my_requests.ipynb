```python
import requests
from IPython.display import HTML
```

```python
# Faz a decodificacao do conteudo para o encoding correto
# Exemplo: codificacao UTF-8: https://pt.wikipedia.org/wiki/UTF-8
def decodifica_conteudo(conteudo):
  try:
      conteudo_str = conteudo.decode('utf-8')
  except UnicodeDecodeError as e:
      try:
        conteudo_str = conteudo.decode('latin-1')
      except UnicodeDecodeError as e:
        try:
          conteudo_str = conteudo.decode('ISO-8859-1')
        except UnicodeDecodeError as e:
          print(f"UnicodeDecodeError: {e}")
          conteudo_str = None

  return conteudo_str
```

```python
# Pesquisa um termo no www.google.com
# https://www.google.com/search?q="Armando Soares Sousa"

termo_consulta = "Armando Soares Sousa"
URL_CONSULTA = f"https://www.google.com/search?q={termo_consulta}"

print("Consulta: ")
print(URL_CONSULTA)

try:
  print("Aguarde...")
  minha_consulta = requests.get(URL_CONSULTA)
  minha_consulta.raise_for_status()
  if minha_consulta.status_code==200:
    conteudo = minha_consulta.content
    conteudo_str = decodifica_conteudo(conteudo)
    if conteudo:
      print(conteudo_str)
    else:
      print("Conteudo incompatível para exibição")
except requests.exceptions.RequestException as e:
  print(f"Erro na requisção: {e}")
except Exception as e:

  print(f"Erro na exibição do conteúdo: {e}")
```
