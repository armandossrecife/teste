## Funções

```python
# importing required modules
from PyPDF2 import PdfReader
import re
import pandas as pd

def get_lista_paginas(qtd_paginas):
  lista_paginas = []
  for i in range(qtd_paginas):
    page = reader.pages[i]
    text = page.extract_text()
    lista_paginas.append(text)

  return lista_paginas

def extract_names_from_text(text):
  # Extract all upper case characters between "- " and "digits"
  list_names = re.findall(r"- ([A-Z ]*)[0-9]+", text)

  return list_names
```

## Ler alunos do arquivo .pdf

```python
lista_alunos_from_pdf = []

# creating a pdf reader object
reader = PdfReader('ListaEleicao.pdf')

qtd_paginas = len(reader.pages)
# printing number of pages in pdf file
print(f'Qtd páginas: {qtd_paginas}')

lista_paginas = get_lista_paginas(qtd_paginas)
for i,pagina in enumerate(lista_paginas):
  print(f'Página: {i+1}')
  alunos_por_pagina = extract_names_from_text(pagina)
  print(alunos_por_pagina)
  for aluno in alunos_por_pagina:
    lista_alunos_from_pdf.append(aluno)
```

```python
print('Alunos extraídos do pdf: ')
print(len(lista_alunos_from_pdf), lista_alunos_from_pdf)
```

## Lista de alunos ativos via SIGAA

```python
alunos_computacao = pd.read_html('https://www.sigaa.ufpi.br/sigaa/public/curso/alunos.jsf?lc=pt_BR&id=74268')
df_alunos_computacao_2024 = alunos_computacao[0][['Matrícula', 'Aluno']]
df_alunos_computacao_2024.drop(df_alunos_computacao_2024.tail(1).index,inplace=True) # drop last n rows
df_alunos_computacao_2024
```

```python
print('Alunos extraídos da URL pública do SIGAA: ')
lista_alunos_from_sigaa_publico = df_alunos_computacao_2024.Aluno.to_list()
print(f'Qtd: {len(lista_alunos_from_sigaa_publico)}, {lista_alunos_from_sigaa_publico}')
```

## Exporta relação de alunos para .csv e .xlsx

```python
df_alunos_computacao_2024.to_csv('alunos_computacao_ativos_2024.csv', index=False)
df_alunos_computacao_2024.to_excel('alunos_computacao_ativos_2024.xlsx', index=False)
```
