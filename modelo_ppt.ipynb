```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN

def create_soberania_presentation():
    # Create a presentation object
    prs = Presentation()

    # Slide 1: Cover
    slide_layout = prs.slide_layouts[0]  # Title slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "Portal SoberanIA-UFPI"
    title.text_frame.paragraphs[0].font.bold = True
    title.text_frame.paragraphs[0].font.size = Pt(44)

    subtitle.text = "Gestão Colaborativa de Projetos\nOrganização, Transparência e Produtividade no GitHub"
    subtitle.text_frame.paragraphs[0].font.size = Pt(24)

    # Add a note about the image
    notes_slide = slide.notes_slide
    notes_slide.notes_text_frame.text = "Adicionar logo da organização e ícones de GitHub/gestão como imagem de fundo"

    # Slide 2: Objectives
    slide_layout = prs.slide_layouts[1]  # Title and content
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Objetivos do Portal"
    content.text = ("- Centralizar a gestão das equipes (Pesquisa e Desenvolvimento)\n"
                   "- Facilitar comunicação com parceiros (DW/GAIA, PIT)\n"
                   "- Garantir rastreabilidade de atividades e documentos\n\n"
                   '"Um único ambiente para pesquisa, código e gestão."')

    # Slide 3: Why GitHub?
    slide_layout = prs.slide_layouts[1]
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Por Que GitHub?"
    content.text = ("- Versionamento: Controle histórico de código e documentos\n"
                   "- Colaboração: Issues, projetos Kanban e revisão de código\n"
                   "- Automação: CI/CD com GitHub Actions\n\n"
                   "Nota: Adicionar print da organização no GitHub ou fluxo de trabalho")

    # Slide 4: Key Benefits
    slide_layout = prs.slide_layouts[1]
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Benefícios Chave"

    # Create a table
    rows = 4
    cols = 2
    left = Inches(1.0)
    top = Inches(1.5)
    width = Inches(8.0)
    height = Inches(4.0)

    table = slide.shapes.add_table(rows, cols, left, top, width, height).table

    # Set column widths
    table.columns[0].width = Inches(3.0)
    table.columns[1].width = Inches(5.0)

    # Write headers
    table.cell(0, 0).text = "Área"
    table.cell(0, 1).text = "Vantagem"

    # Write content
    table.cell(1, 0).text = "Documentação"
    table.cell(1, 1).text = "Markdown + páginas interlinkadas"

    table.cell(2, 0).text = "Tarefas"
    table.cell(2, 1).text = "Issues + GitProjects integrados"

    table.cell(3, 0).text = "Código"
    table.cell(3, 1).text = "Code Review + CI/CD automatizado"

    # Format table
    for cell in table.iter_cells():
        cell.text_frame.paragraphs[0].font.size = Pt(14)
        cell.text_frame.paragraphs[0].alignment = PP_ALIGN.LEFT

    # Slide 5: Repository Structure
    slide_layout = prs.slide_layouts[1]
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.placeholders[1]

    title.text = "Estrutura dos Repositórios"

    # Create a table
    rows = 4
    cols = 4
    left = Inches(0.5)
    top = Inches(1.5)
    width = Inches(9.0)
    height = Inches(4.0)

    table = slide.shapes.add_table(rows, cols, left, top, width, height).table

    # Set column widths
    table.columns[0].width = Inches(2.5)
    table.columns[1].width = Inches(2.0)
    table.columns[2].width = Inches(3.0)
    table.columns[3].width = Inches(1.5)

    # Write headers
    table.cell(0, 0).text = "Repositório"
    table.cell(0, 1).text = "Admin"
    table.cell(0, 2).text = "Função"
    table.cell(0, 3).text = "Link"

    # Write content
    table.cell(1, 0).text = "Gestão"
    table.cell(1, 1).text = "Pedro de Alcântara"
    table.cell(1, 2).text = "Processos e Kanban estratégico"
    table.cell(1, 3).text = "Link"

    table.cell(2, 0).text = "Pesquisa"
    table.cell(2, 1).text = "Raimundo Moura"
    table.cell(2, 2).text = "Publicação de artigos/dados"
    table.cell(2, 3).text = "Link"

    table.cell(3, 0).text = "Desenvolvimento"
    table.cell(3, 1).text = "Armando Soares"
    table.cell(3, 2).text = "Código-fonte e pipelines"
    table.cell(3, 3).text = "Link"

    # Format table
    for cell in table.iter_cells():
        cell.text_frame.paragraphs[0].font.size = Pt(12)
        cell.text_frame.paragraphs[0].alignment = PP_ALIGN.LEFT

    # Add footer note
    notes_slide = slide.notes_slide
    notes_slide.notes_text_frame.text = "Todos os colaboradores terão acesso conforme sua equipe."

    # Continue with other slides following the same pattern...
    # (Slide 6: Workflow, Slide 7: How to Contribute, etc.)

    # Save the presentation
    prs.save("SoberanIA-UFPI_Portal.pptx")
    print("Apresentação criada com sucesso: SoberanIA-UFPI_Portal.pptx")

create_soberania_presentation()
```
