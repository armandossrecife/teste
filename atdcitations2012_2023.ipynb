```python
import pandas as pd
```

```python
df_atd_citations = pd.read_csv('atdcitations.csv')
df_atd_citations
```

```python

df_atd_citations.info()
```

```python
my_columns = ['Title', 'Authors', 'Year', 'Publication']
df_my_data = df_atd_citations[my_columns]

df_my_data_valids = df_my_data.dropna()
df_my_data_valids['Year'] = df_my_data_valids['Year'].astype(int)
df_my_data_valids = df_my_data_valids.sort_values(by='Year')
```

```python
df_my_data_valids
```

```python
df_my_data_valids.to_csv('atdcitations2012-2023.csv', index=False)
```

```python
df_my_data_valids_links = pd.read_excel('atdcitations2012-2023.xlsx')
df_my_data_valids_links
```

```python
element_counts_by_year = df_my_data_valids_links['Year'].to_list()
```

```python
from collections import Counter

element_counts = Counter(element_counts_by_year)
```

```python
for k, v in element_counts.items():
  print(k, v)
```
