## Extrai dados básicos de uma imagem

.pgn ou .jpg

```python
def extrac_image_info(image_path):
    image = Image.open(image_path)

    width, height = image.size
    filename = os.path.basename(image_path)
    size = os.path.getsize(image_path)
    format = image.format

    image_info =  {
        'filename': filename,
        'width': width,
        'height': height,
        'size': size,
        'format': format
    }
    return image_info
```

## Extrai dados exif de uma image .jpeg que tenha tais informações

```python
def extract_exif_image_info(image_path):
    exif_data = {}
    with open(image_path, 'rb') as f:
      try:
        exif_tags = exifread.process_file(f)
        for tag, value in exif_tags.items():
          if value:
            exif_data[tag] = value
        return exif_data
      except Exception as e:  # Catch any exceptions during EXIF processing
        raise ValueError(f"Error processing EXIF data for {image_path}: {str(e)}")
        return None
```

## Exemplos

```python
lista_image_paths = []

current_path = os.getcwd()
print(current_path)

my_path1 = current_path + '/' + 'babydogcat.png'
print(my_path1)
my_path2 = current_path + '/' + 'screen_matrix.jpeg'
print(my_path2)
my_path3 = current_path + '/' + 'DSCN0010.jpg'
print(my_path3)
my_path4 = current_path + '/' + 'HMD_Nokia_8.3_5G.jpg'
print(my_path4)

lista_image_paths.append(my_path1)
lista_image_paths.append(my_path2)
lista_image_paths.append(my_path3)
lista_image_paths.append(my_path4)
```

```python
try:
  for image_path in lista_image_paths:
    image_info = extrac_image_info(image_path)
    print(image_info)
except Exception as e:
  print(f"Error processing image: {str(e)}")
```

```python
image_exif1 = extract_exif_image_info(my_path1)
print(image_exif1)
image_exif2 = extract_exif_image_info(my_path2)
print(image_exif2)
```

```python
image_exif = extract_exif_image_info(my_path3)
print(image_exif)
```

```python
print(my_path3)
for k, v in image_exif.items():
  print(f"{k}: {v}")
```
