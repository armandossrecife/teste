```python
from datasets import load_dataset
from huggingface_hub import login
import concurrent.futures
import os
import time
from google.colab import userdata
```

```python
# Configuration
HF_TOKEN = userdata.get('HF_TOKEN')
dataset_name = "ErikCikalleshi/new_york_times_news_2000_2007"
dataset_config = "?"  # Specific Wikipedia snapshot
output_dir = "dataset_download"
num_threads = 10  # Number of parallel workers

# Create output directory
os.makedirs(output_dir, exist_ok=True)
login(HF_TOKEN)
```

```python
def download_shard(shard_idx, num_shards):
    try:
        print(f"Downloading shard {shard_idx}/{num_shards}")
        # Load a single shard (works for Parquet-based datasets)
        dataset = load_dataset(
            dataset_name,
            split="train",
            streaming=False,  # Disable streaming for direct Parquet access
        )

        # Save the shard (since the dataset is already sharded)
        output_file = os.path.join(output_dir, f"shard_{shard_idx}.parquet")
        dataset.to_parquet(output_file)
        return True
    except Exception as e:
        print(f"Error in shard {shard_idx}: {str(e)}")
        return False
```

```python
# Get total number of shards (approximate)
try:
    dataset_info = load_dataset(dataset_name, dataset_config, split="train", streaming=True)
    num_shards = dataset_info.num_shards if hasattr(dataset_info, 'num_shards') else 100  # Fallback
except:
    num_shards = 100  # Default if detection fails
```

```python
print(f"Downloading {num_shards} shards using {num_threads} workers")

# Parallel download with ThreadPoolExecutor
start_time = time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
    futures = [executor.submit(download_shard, i, num_shards) for i in range(num_shards)]
    results = [future.result() for future in concurrent.futures.as_completed(futures)]

print(f"Download completed in {time.time() - start_time:.2f} seconds")
print(f"Successfully downloaded {sum(results)}/{num_shards} shards")
```
