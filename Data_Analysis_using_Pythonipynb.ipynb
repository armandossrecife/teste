# 1. Import libs and load data

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Genre': ['RPG', 'Strategy', 'FPS', 'Adventure', 'Simulation', 'RPG', 'Strategy'],
    'Average Playtime (hours)': [50, 40, 25, 30, 35, 60, 55]
}

df = pd.DataFrame(data)
```

```python
df
```

# 2. Data Exploration

```python
# Get some summary information about the data
print(df.info())

# Get descriptive statistics of playtime
playtime_stats = df['Average Playtime (hours)'].describe()
print(f"{round(playtime_stats, 2)}")
```

# 3. Data Vizualization

```python
import matplotlib.pyplot as plt

# Create a bar chart to visualize average playtime by genre
plt.xlabel('Video Game Genre')
plt.ylabel('Average Playtime (hours)')
plt.title('Average Playtime by Video Game Genre')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.bar(df['Genre'], df['Average Playtime (hours)'])
plt.show()
```
