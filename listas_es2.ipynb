```python
lista_alunos = []
with open('alunos_es2_20232.txt', mode='r') as f_alunos:
  for line in f_alunos:
    line = line.strip()
    aluno_temp = line.split(' ')
    matricula = aluno_temp[0]
    nome = ' '.join(aluno_temp[1:])
    aluno = matricula, nome
    lista_alunos.append(aluno)
```

```python
lista_equipes = []
dict_equipes = {}
with open('alunos_equipes.txt',mode='r') as f_equipes:
  for line in f_equipes:
    line = line.split(',')
    matricula = line[0]
    nome = line[1]
    equipe = line[2].strip()
    dict_equipes[matricula] = nome, equipe
```

```python
dict_tutoriais_requisitos = {
    'equipe1': [('tutoriais',4), ('requisitos', 3)],
	'equipe2': [('tutoriais',3.5), ('requisitos',3)],
	'equipe3': [('tutoriais',4), ('requisitos',3)],
	'equipe4': [('tutoriais',3), ('requisitos',3)],
	'equipe5': [('tutoriais',4), ('requisitos', 3)],
	'equipe6': [('tutoriais',4), ('requisitos', 3)],
	'equipe7': [('tutoriais',4), ('requisitos',3)]
}
```

```python
lista_label_l1, lista_label_l2, lista_label_l3, lista_label_equipe, lista_label_tutoriais, lista_label_requisitos = [], [], [], [], [], []

for k, v in dict_alunos_es2.items():
  if v in lista1_nomes:
    lista_label_l1.append(1)
  else:
    lista_label_l1.append(0)

  if v in lista2_nomes:
    lista_label_l2.append(1)
  else:
    lista_label_l2.append(0)

  if v in lista3_nomes:
    lista_label_l3.append(1)
  else:
    lista_label_l3.append(0)

  my_tuple = dict_equipes[k]
  equipe = my_tuple[1]
  lista_label_equipe.append(equipe)

  dict_tutoriais_requisitos[equipe]
  dados = dict_tutoriais_requisitos[equipe]
  tutoriais = dados[0]
  requisitos = dados[1]

  lista_label_tutoriais.append(tutoriais[1])
  lista_label_requisitos.append(requisitos[1])
```
