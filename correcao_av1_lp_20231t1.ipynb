# Pre-processamento

```python
def salva_alunos_em_lista(nome_arquivo):
  alunos = []
  with open(nome_arquivo, mode='r') as f_temp:
    for linha in f_temp:
      alunos.append(linha.strip())
  return alunos

def remove_tabulacao(conteudo):
  conteudo = conteudo.strip()
  return conteudo
```

```python
import pandas as pd

df_alunos_lp_temp = pd.read_csv('alunoslp.csv')
df_alunos_lp_temp['nome'].apply(lambda x: remove_tabulacao(x))
lista_nomes = df_alunos_lp_temp['nome'].apply(lambda x: remove_tabulacao(x)).to_list()
lista_nomes
```

```python
alunos_l1, alunos_l2, alunos_l3, alunos_l4 = [], [], [], []
alunos_l1 = salva_alunos_em_lista(nome_arquivo='l1.txt')
alunos_l2 = salva_alunos_em_lista(nome_arquivo='l2.txt')
alunos_l3 = salva_alunos_em_lista(nome_arquivo='l3.txt')
alunos_l4 = salva_alunos_em_lista(nome_arquivo='l4.txt')
print(len(alunos_l1))
print(len(alunos_l2))
print(len(alunos_l3))
print(len(alunos_l3))
```

```python
fez_l1, fez_l2, fez_l3, fez_l4 = ([], [], [], [])
for nome in lista_nomes:
  if nome in alunos_l1:
    fez_l1.append(1)
  else:
    fez_l1.append(0)
  if nome in alunos_l2:
    fez_l2.append(1)
  else:
    fez_l2.append(0)
  if nome in alunos_l3:
    fez_l3.append(1)
  else:
    fez_l3.append(0)
  if nome in alunos_l4:
    fez_l4.append(1)
  else:
    fez_l4.append(0)
```

```python
dic_alunos_lp = {
    'nome':lista_nomes,
    'fez_l1':fez_l1,
    'fez_l2':fez_l2,
    'fez_l3':fez_l3,
    'fez_l4':fez_l4,
}
```

```python
dic_alunos_lp
```

```python
df_alunos_lp_av1 = pd.DataFrame(dic_alunos_lp)
df_alunos_lp_av1['soma'] = df_alunos_lp_av1['fez_l1'] + df_alunos_lp_av1['fez_l2'] + df_alunos_lp_av1['fez_l3'] + df_alunos_lp_av1['fez_l4']
df_alunos_lp_av1
```

```python
df_alunos_lp_av1.to_csv('alunos_av1.csv', index=False)
```
