```python
df_alunos_pe = pd.read_csv('alunos_pe.csv')
df_alunos_pe.columns = ['id', 'matricula', 'nome']
lista_nomes = df_alunos_pe.nome.tolist()

duplas_espontaneas = [('IAN SURIEL ALVES FERREIRA','FRANCISCA TAIANE PEREIRA MOURAO'),
('JOAO HENRIQUE LIMA OLIVEIRA','ARTUR DE ARAUJO CASTRO'),
('VICTOR GABRIEL RIBEIRO COELHO','YAGO LOHANN LIMA MAGALHAES'),
('VANDERSON DE PAULA DA SILVA','EMERSON SOUSA LIMA'),
('RIAN GABRIEL MORAES DE CARVALHO','PAULO HENRIQUE DA SILVA JUNIOR'),
('MARIA CLARA DA COSTA SOUSA BONFIM','ANA KAROLINE SILVA MARTINS'),
('ALINE CRISLAINY BARBOSA DA SILVA SOUSA','ADEILTON JUSCELINO DA SILVA'),
('SAMUEL BEZERRA DE MENEZES DA SILVA','JOANA DARC CAMPELO FERNANDES'),
('JHUAN PABLO DE SOUSA AQUINO','TIPHANY NATSU NISHIMURA ANDRADE ALVES'),
('KLEMESON LUCAS DE SOUSA NEVES','NAYWAN RAPHAEL CALACA DE OLIVEIRA'),
('MOISES DA SILVA BASTOS','TERESA CRISTINA SOARES DE CARVALHO')]

lista_nomes_ok = []
for dupla in duplas_espontaneas:
  lista_nomes_ok.append(dupla[0])
  lista_nomes_ok.append(dupla[1])

lista_nomes_para_sorteio = []
for nome in lista_nomes:
  if nome not in lista_nomes_ok:
    lista_nomes_para_sorteio.append(nome)

  
novas_duplas = []
novas_duplas.append((lista_nomes_para_sorteio[0], lista_nomes_para_sorteio[1]))
novas_duplas.append((lista_nomes_para_sorteio[2], lista_nomes_para_sorteio[3]))
novas_duplas.append((lista_nomes_para_sorteio[4], lista_nomes_para_sorteio[5]))
novas_duplas.append((lista_nomes_para_sorteio[6], lista_nomes_para_sorteio[7]))
novas_duplas.append((lista_nomes_para_sorteio[8], lista_nomes_para_sorteio[9]))
novas_duplas.append((lista_nomes_para_sorteio[10], lista_nomes_para_sorteio[11]))

duplas_projeto_final = duplas_espontaneas + novas_duplas

dict_duplas_temp = {}

lista_temp = []
for i, each in enumerate(duplas_projeto_final):
  dupla = 'dupla ' + str(i+1)
  lista_temp.append(each)
  dict_duplas_temp[dupla] = lista_temp
  lista_temp = []

lista_k = []
lista_v = []
for k, v in dict_duplas_temp.items():
  lista_k.append(k)
  lista_v.append((v[0][0], v[0][1]))

dict_duplas_pe = {
    'Duplas':lista_k,
    'Nomes':lista_v
}

df_duplas_pe = pd.DataFrame(dict_duplas_pe)
df_duplas_pe
```
