```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.gapminder()
df.head()
```

```python
df.info()
```

```python
my_data = df.query("year==2007")
my_data.head()
```

```python
fig = px.scatter(my_data,
                 x="gdpPercap", y="lifeExp", size="pop",
                 color="continent", hover_name="country",
                 log_x=True, size_max=60)
```

```python
fig
```

```python
df = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
```

```python
fig
```

```python
import numpy as np

df = px.data.gapminder().query("year == 2007")
fig = px.treemap(df, path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
```

```python
df_capitais_brasil = pd.read_csv('capitais_brasil.csv')
df_capitais_brasil
```

```python
import random

def gerar_numero_aleatorio():
    return random.randint(30, 3000)

def gerar_peak_hour():
    return random.randint(0, 23)
```

```python
# Create a new column with random car_hours
df_capitais_brasil['car_hours'] = df_capitais_brasil.apply(lambda row: gerar_numero_aleatorio(), axis=1)
df_capitais_brasil['peak_hour'] = df_capitais_brasil.apply(lambda row: gerar_peak_hour(), axis=1)
```

```python
df_capitais_brasil['centroid_lat'] = df_capitais_brasil['Latitude']
df_capitais_brasil['centroid_lon'] = df_capitais_brasil['Longitude']
```

```python
df_capitais_brasil
```

```python
fig = px.scatter_mapbox(df_capitais_brasil, lat="Latitude", lon="Longitude", color="peak_hour", size="car_hours",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10,
                  mapbox_style="carto-positron")
```

```python
fig
```
