## Dada a seguinte função:

$$y = x^2 - 1$$

### Representação em Python

```python
y = x**2 - 1
```

## Criando o gráfico via matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values
y = x**2 - 1

# Create the plot
plt.plot(x, y)

# Add the horizontal line at y = 0
plt.axhline(y=0, color='red', linestyle='--', label='y = 0')

# Set the title and labels
plt.title("Plot of x^2 - 1")
plt.xlabel("x")
plt.ylabel("y")

# Show the plot
plt.show()
````

## Encontrando a solução da função

Usando o symply

https://docs.sympy.org/latest/index.html

via solvers

https://docs.sympy.org/latest/modules/solvers/solvers.html

````python
from sympy.solvers import solve
from sympy import Symbol

x = Symbol('x')
solutions = solve(x**2 - 1, x)

print("The solutions are:")
for solution in solutions:
    print(solution)
````

````python

from sympy.solvers import solve
from sympy import Symbol
import numpy as np

x = Symbol('x')
solutions = solve(x**2 - 1, x)
````

````python
print(*solutions)
````

````python
[0, *solutions]
````

````python
# Evaluate the function at the critical points and endpoints
y_values = [x**2 - 1 for x in [0, *solutions]]
````

````python
print(y_values)
````

````python
# Find the minimum value
min_value = min(y_values)

print("The minimum value of y = x^2 - 1 is:", min_value)
````

````python
# Define the x values
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values
y = x**2 - 1

# Create the plot
plt.plot(x, y)

# Add the horizontal line at y = 0
plt.axhline(y=0, color='black', linestyle='-', label='y = 0')
# Add the vertical line at x = -1
plt.axvline(x=-1, color='red', linestyle='--', label='x = -1')
# Add the vertical line at x = 1
plt.axvline(x=1, color='red', linestyle='--', label='x = 1')
# Add the horizontal line at y = -1
plt.axhline(y=-1, color='red', linestyle='--', label='y = -1')
# Add the vertical line at x = -1

# Set the title and labels
plt.title("Plot of x^2 - 1")
plt.xlabel("x")
plt.ylabel("y")

# Set the x-axis ticks
plt.xticks([-1, 1])

# Set the y-axis ticks
plt.yticks([min_value])

# Add a legend
plt.legend()
# Show the plot
plt.show()
````
