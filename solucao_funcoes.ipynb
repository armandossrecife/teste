````python
from sympy.solvers import solve
from sympy import Symbol
import numpy as np

x = Symbol('x')
solutions = solve(x**2 - 1, x)
````

````python
print(*solutions)
````

````python
[0, *solutions]
````

````python
# Evaluate the function at the critical points and endpoints
y_values = [x**2 - 1 for x in [0, *solutions]]
````

````python
print(y_values)
````

````python
# Find the minimum value
min_value = min(y_values)

print("The minimum value of y = x^2 - 1 is:", min_value)
````

## Encontrando o Mínimo de uma Função Quadrática Usando Cálculo Diferencial

**O que é o cálculo diferencial?**

O cálculo diferencial é uma ferramenta matemática que nos permite estudar como as funções variam. Uma das suas principais aplicações é encontrar os pontos máximos e mínimos de uma função.

**Por que usar o cálculo diferencial?**

Ao encontrar a derivada de uma função e igualá-la a zero, estamos, na verdade, procurando os pontos onde a inclinação da tangente à curva é nula. Em um gráfico, esses pontos correspondem aos picos (máximos) e vales (mínimos) da função.

**Resolvendo o problema: f(x) = x² - 1**

1. **Calcular a derivada:**
   A derivada de f(x) = x² - 1 é f'(x) = 2x.

2. **Igualar a derivada a zero:**
   2x = 0

3. **Resolver a equação:**
   x = 0

4. **Analisar o resultado:**
   Encontramos um ponto crítico em x = 0. Para verificar se este ponto corresponde a um mínimo, podemos analisar a concavidade da função neste ponto. A segunda derivada de f(x) é f''(x) = 2, que é sempre positiva. Isso indica que a função é côncava para cima em todo o seu domínio, ou seja, o ponto x = 0 é um ponto de mínimo global.

**Conclusão:**

O valor mínimo da função f(x) = x² - 1 ocorre em x = 0. Substituindo x = 0 na função original, encontramos o valor mínimo: f(0) = -1.

**Em resumo:**

Ao utilizar o cálculo diferencial, conseguimos encontrar o ponto exato onde a função atinge seu menor valor. Essa técnica é fundamental em diversas áreas, como engenharia, economia e física, para otimizar processos e resolver problemas.

**Observações:**

* **Pontos críticos:** Os pontos onde a derivada se anula ou não existe são chamados de pontos críticos. Nem todos os pontos críticos correspondem a máximos ou mínimos.
* **Concavidade:** A segunda derivada nos informa sobre a concavidade da função. Se a segunda derivada é positiva, a função é côncava para cima (forma de um "U") e o ponto crítico é um mínimo. Se a segunda derivada é negativa, a função é côncava para baixo (forma de um "∩") e o ponto crítico é um máximo.

**Outras aplicações:**

* **Otimização de custos:** Encontrar a quantidade de produção que minimiza o custo de produção.
* **Máxima altura de um projétil:** Encontrar o instante em que um projétil atinge sua altura máxima.
* **Curvas de aprendizado:** Modelar como a eficiência de uma pessoa aumenta com a prática.
