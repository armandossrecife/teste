# Dataframes

```python
df_alunos = pandas.read_csv('alunostes20241.csv')
df_alunos = df_alunos[['matricula', 'nome']]
df_alunos.head()
```

```python
df_listas = pandas.read_csv('alunos_listas_tes20241.csv')
df_listas.head()
```

```python
df_aws_av1 = pandas.read_csv('aws3_av1.csv', sep=';')
```

```python
df_aws_av1_by_name = df_aws_av1.sort_values(by=['Nome'])
df_aws_av1_by_name.head()
```

```python
colunas_uteis = ['Matricula',	'Nome', 'Final Points', 'Final Score']
df_aws_av1_by_name = df_aws_av1_by_name[colunas_uteis]
df_aws_av1_by_points = df_aws_av1_by_name.sort_values(by=['Final Points'], ascending=False)
df_aws_av1_by_points
```

```python
colunas_uteis
```

```python
df_aws_av1_by_name = df_aws_av1_by_name[colunas_uteis]
df_aws_av1_by_name = df_aws_av1_by_name.rename(columns={'Matricula': 'matricula', 'Nome': 'nome', 'Final Points': 'final_points', 'Final Score': 'final_score'})
df_aws_av1_by_name
```

```python
# prompt: how to merge the following dataframes: df_aws_av1_by_name and df_listas?

df_merged = pandas.merge(left=df_aws_av1_by_name, right=df_listas, on='matricula', how='left')
df_merged
```

```python
df_merged.columns
```

```python
novas_colunas = ['matricula', 'nome_y', 'fez_lista1', 'fez_lista2', 'fez_lista3', 'fez_lista4', 'final_points', 'final_score']
df_alunos_av1 = df_merged[novas_colunas]
df_alunos_av1
```

```python
df_alunos_av1.to_csv('alunos_av1.csv', index=False)
```
