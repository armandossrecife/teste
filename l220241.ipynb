```python
def menu():
    # Exibe o menu de opções
    print("\nMenu de Gerenciamento de Alunos:")
    print("1. Inserir Aluno")
    print("2. Buscar Aluno")
    print("3. Exibir Lista de Alunos (Ordenada)")
    print("4. Remover Aluno")
    print("5. Sair")

def insere_aluno(lista, nome):
  lista.append(nome)

def busca_aluno(lista, nome):
  if nome in lista:
    return True
  else:
    return False

def exibir_lista_alunos(lista):
      # Exibir lista ordenada
      lista.sort()
      for aluno in lista:
        print(f"- {aluno}")

def remover_aluno(lista, nome):
  if nome in lista:
    lista.remove(nome)
    print(f"Aluno {nome} removido com sucesso!")
  else:
    print(f"Aluno {nome} não encontrado na lista.")

def gerenciar_lista_alunos():
  """
  Gerencia uma lista de nomes de alunos, permitindo operações de inserção, busca, ordenação e remoção.
  Argumentos:
    None
  Retorna:
    None
  """
  # Inicializa a lista de alunos vazia
  alunos = []

  while True:
    menu()
    # Solicita a opção do usuário
    opcao = input("Digite a opção desejada: ")

    if opcao == "1":
      # Inserir aluno
      nome_aluno = input("Digite o nome do aluno: ")
      insere_aluno(alunos, nome_aluno)
      print(f"Aluno {nome_aluno} inserido com sucesso!")

    elif opcao == "2":
      # Buscar aluno
      nome_busca = input("Digite o nome do aluno a buscar: ")
      if busca_aluno(alunos, nome_busca):
        print(f"Aluno {nome_busca} encontrado na lista.")
      else:
        print(f"Aluno {nome_busca} não encontrado na lista.")

    elif opcao == "3":
      exibir_lista_alunos(alunos)
    elif opcao == "4":
      # Remover aluno
      nome_remocao = input("Digite o nome do aluno a remover: ")
      remover_aluno(alunos, nome_remocao)

    elif opcao == "5":
      # Sair do programa
      print("Saindo do programa de gerenciamento de alunos...")
      break

    else:
      # Opção inválida
      print("Opção inválida. Tente novamente.")

gerenciar_lista_alunos()
```

```python
def menu_rotas():
    # Exibe o menu de opções
    print("\nMenu de Gerenciamento de Rota GPS:")
    print("1. Adicionar Ponto GPS")
    print("2. Remover Ponto GPS")
    print("3. Exibir Rota Atual")
    print("4. Sair")

def adicionar_ponto_gps(latitude, longitude, rota):
    novo_ponto = (latitude, longitude)  # Cria uma tupla com as coordenadas
    rota.append(novo_ponto)  # Adiciona o ponto à rota

def remover_ponto_gps(rota):
  if not rota:
    print("A rota está vazia. Nenhum ponto a ser removido.")
    return

  # Exibe os pontos da rota numerados
  for indice, ponto in enumerate(rota):
    print(f"{indice + 1}. {ponto}")

    # Solicita o índice do ponto a remover
    indice_remocao = int(input("Digite o índice do ponto a remover: ")) - 1

    if 0 <= indice_remocao < len(rota):
      del rota[indice_remocao]  # Remove o ponto da rota
      print(f"Ponto GPS removido com sucesso!")
    else:
      print("Índice inválido. Tente novamente.")

def listar_pontos_gps(rota):
  if not rota:
    print("A rota está vazia. Nenhum ponto a ser exibido.")
    return

  # Exibe os pontos da rota numerados
  print(rota)

def gerenciar_rota_gps():
  """
  Gerencia uma lista de tuplas de pontos GPS que representam uma rota, permitindo operações de inserção e remoção de pontos.

  Argumentos:
    None

  Retorna:
    None
  """

  # Inicializa a lista de pontos GPS vazia
  rota = []

  while True:
    menu_rotas()
    # Solicita a opção do usuário
    opcao = input("Digite a opção desejada: ")
    if opcao == "1":
      # Adicionar ponto GPS
      latitude = float(input("Digite a latitude: "))
      longitude = float(input("Digite a longitude: "))
      adicionar_ponto_gps(latitude, longitude, rota)
      print(f"Ponto GPS ({latitude}, {longitude}) adicionado com sucesso!")
    elif opcao == "2":
      # Remover ponto GPS
      remover_ponto_gps(rota)
    elif opcao == "3":
      listar_pontos_gps(rota)
    elif opcao == "4":
      # Sair do programa
      print("Saindo do programa de gerenciamento de rota GPS...")
      break
    else:
      # Opção inválida
      print("Opção inválida. Tente novamente.")

gerenciar_rota_gps()
```

```python
def gerenciar_conjuntos_cores():
  """
  Gerencia três conjuntos de cores, permitindo operações de união, intersecção e diferença.

  Argumentos:
    None

  Retorna:
    None
  """

  # Inicializa os conjuntos vazios
  conjunto1 = set()
  conjunto2 = set()
  conjunto3 = set()

  # Adiciona algumas cores aos conjuntos
  conjunto1.add("azul")
  conjunto1.add("verde")
  conjunto1.add("vermelho")

  conjunto2.add("amarelo")
  conjunto2.add("verde")
  conjunto2.add("laranja")

  conjunto3.add("rosa")
  conjunto3.add("azul")
  conjunto3.add("preto")

  while True:
    # Exibe o menu de opções
    print("\nMenu de Gerenciamento de Conjuntos de Cores:")
    print("1. União (Conjunto 1 e Conjunto 2)")
    print("2. União (Conjunto 1 e Conjunto 3)")
    print("3. União (Conjunto 2 e Conjunto 3)")
    print("4. Intersecção (Conjunto 1 e Conjunto 2)")
    print("5. Intersecção (Conjunto 1 e Conjunto 3)")
    print("6. Intersecção (Conjunto 2 e Conjunto 3)")
    print("7. Diferença (Conjunto 1 - Conjunto 2)")
    print("8. Diferença (Conjunto 1 - Conjunto 3)")
    print("9. Diferença (Conjunto 2 - Conjunto 3)")
    print("10. Sair")

    # Solicita a opção do usuário
    opcao = input("Digite a opção desejada: ")

    if opcao == "1":
      # União (Conjunto 1 e Conjunto 2)
      resultado = conjunto1 | conjunto2
      print(f"União (Conjunto 1 e Conjunto 2): {resultado}")

    elif opcao == "2":
      # União (Conjunto 1 e Conjunto 3)
      resultado = conjunto1 | conjunto3
      print(f"União (Conjunto 1 e Conjunto 3): {resultado}")

    elif opcao == "3":
      # União (Conjunto 2 e Conjunto 3)
      resultado = conjunto2 | conjunto3
      print(f"União (Conjunto 2 e Conjunto 3): {resultado}")

    elif opcao == "4":
      # Intersecção (Conjunto 1 e Conjunto 2)
      resultado = conjunto1 & conjunto2
      print(f"Intersecção (Conjunto 1 e Conjunto 2): {resultado}")

    elif opcao == "5":
      # Intersecção (Conjunto 1 e Conjunto 3)
      resultado = conjunto1 & conjunto3
      print(f"Intersecção (Conjunto 1 e Conjunto 3): {resultado}")

    elif opcao == "6":
      # Intersecção (Conjunto 2 e Conjunto 3)
      resultado = conjunto2 & conjunto3
      print(f"Intersecção (Conjunto 2 e Conjunto 3): {resultado}")

    elif opcao == "7":
      # Diferença (Conjunto 1 - Conjunto 2)
      resultado = conjunto1 - conjunto2
      print(f"Diferença (Conjunto 1 - Conjunto 2): {resultado}")

    elif opcao == "8":
      # Diferença (Conjunto 1 - Conjunto 3)
      resultado = conjunto1 - conjunto3
      print(f"Diferença (Conjunto 1 - Conjunto 3): {resultado}")

    elif opcao == "9":
      # Diferença (Conjunto 2 - Conjunto 3)
      resultado = conjunto2 - conjunto3
      print(f"Diferença (Conjunto 2 - Conjunto 3): {resultado}")

    elif opcao == "10":
      # Sair do programa
      print("Saindo do programa de gerenciamento de conjuntos de cores...")
      break

    else:
      # Opção inválida
      print("Opção inválida. Tente novamente.")


gerenciar_conjuntos_cores()
```
