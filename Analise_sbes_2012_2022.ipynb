```python
import pandas as pd
```

```python
df_sbes_10i = pd.read_csv('sbes2022-dataset.csv')
df_sbes_10i
```

```python
df_sbes_10i.info()
```

```python
my_df = df_sbes_10i.copy()
my_df = my_df[['paper_year', 'paper_title', 'paper_authors', 'paper_language', 'citation_year']]
my_df.head(5)
```

```python
my_df.drop_duplicates(inplace=True)
```

```python
my_df['paper_language'].value_counts()
```

```python
conta_linguagem = my_df['paper_language'].value_counts()
conta_linguagem
```

```python
total_linguagem = conta_linguagem['english'] + conta_linguagem['portuguese']
print(f"Total: {total_linguagem}, inglês: {conta_linguagem['english']}, português: {conta_linguagem['portuguese']}")
print(f"inglês: {round(conta_linguagem['english']/total_linguagem, 2)}, português: {round(conta_linguagem['portuguese']/total_linguagem,2)}")
```
