# Versão Modularizada

```python
# Configuração para exibir todas as colunas no DataFrame (opcional)
pd.set_option('display.max_columns', None)

def load_and_standardize_data(student_file_path, list_files_prefix, assessment_name):
    """
    Carrega os dados dos alunos e das listas de exercícios, padronizando os nomes.

    Args:
        student_file_path (str): Caminho do arquivo CSV com a lista de alunos.
        list_files_prefix (str): Prefixo do caminho das listas (ex: 'av1/lista').
        assessment_name (str): Nome da avaliação (ex: 'av1' ou 'av2').

    Returns:
        tuple: (df_alunos, listas_de_nomes)
            - df_alunos: DataFrame com os alunos.
            - listas_de_nomes: Lista contendo os nomes de cada lista (l1, l2, l3, l4).
    """
    # Carrega a lista de alunos e padroniza a coluna 'Nome'
    df_alunos = pd.read_csv(student_file_path)
    df_alunos = df_alunos.rename(columns={'NOME': 'Nome'})

    # Carrega as 4 listas de exercícios e extrai os nomes
    listas_de_nomes = []
    for i in range(1, 5):
        list_file = f"{list_files_prefix}{i}.csv"
        df_list = pd.read_csv(list_file)
        listas_de_nomes.append(df_list['Nome'].tolist())

    return df_alunos, listas_de_nomes

def update_list_status(df_alunos, listas_de_nomes, assessment_name):
    """
    Atualiza o DataFrame com os status de entrega das listas (0 ou 1).

    Args:
        df_alunos (pd.DataFrame): DataFrame com os alunos.
        listas_de_nomes (list): Lista contendo os nomes de cada lista.
        assessment_name (str): Nome da avaliação (ex: 'av1' ou 'av2').

    Returns:
        pd.DataFrame: DataFrame atualizado com as colunas l1, l2, l3, l4.
    """
    # Cria colunas para cada lista (l1, l2, l3, l4)
    for i in range(1, 5):
        col_name = f"l{i}"
        df_alunos[col_name] = 0

    # Atualiza o status (1 se o aluno entregou a lista)
    for i, lista in enumerate(listas_de_nomes, 1):
        col_name = f"l{i}"
        df_alunos[col_name] = df_alunos['Nome'].apply(lambda x: 1 if x in lista else 0)

    return df_alunos

def calculate_grade(df_alunos, assessment_name):
    """
    Calcula a nota da avaliação baseada nas listas entregues.

    Args:
        df_alunos (pd.DataFrame): DataFrame com os alunos.
        assessment_name (str): Nome da avaliação (ex: 'av1' ou 'av2').

    Returns:
        pd.DataFrame: DataFrame com as colunas de nota calculadas.
    """
    # Soma as listas entregues (av1 = l1 + l2 + l3 + l4)
    df_alunos[f"{assessment_name}"] = df_alunos[['l1', 'l2', 'l3', 'l4']].sum(axis=1)

    # Calcula a nota final (n1 = av1 * 2.5)
    df_alunos[f"n{assessment_name[-1]}"] = df_alunos[f"{assessment_name}"] * 2.5

    return df_alunos

def process_assessment(student_file_path, list_files_prefix, assessment_name, output_file):
    """
    Processa uma avaliação completa (carrega dados, atualiza status, calcula nota e exporta).

    Args:
        student_file_path (str): Caminho do arquivo CSV com a lista de alunos.
        list_files_prefix (str): Prefixo do caminho das listas (ex: 'av1/lista').
        assessment_name (str): Nome da avaliação (ex: 'av1' ou 'av2').
        output_file (str): Nome do arquivo Excel de saída.
    """
    # Carrega e padroniza os dados
    df_alunos, listas_de_nomes = load_and_standardize_data(student_file_path, list_files_prefix, assessment_name)

    # Atualiza o status das listas
    df_alunos = update_list_status(df_alunos, listas_de_nomes, assessment_name)

    # Calcula a nota
    df_alunos = calculate_grade(df_alunos, assessment_name)

    # Remove NaN e exporta para Excel
    df_alunos = df_alunos.fillna(0)
    df_alunos.to_excel(output_file, index=False)

    print(f"Processamento concluído. Arquivo salvo em: {output_file}")
```

```python
# Dados da primeira avaliação (av1)
process_assessment(
    student_file_path='alunos_pe_20251.csv',
    list_files_prefix='av1/lista',
    assessment_name='av1',
    output_file='av1.xlsx'
)
```

```python
# Dados da segunda avaliação (av2)
process_assessment(
    student_file_path='alunos_pe_20251.csv',
    list_files_prefix='av2/lista',
    assessment_name='av2',
    output_file='av2.xlsx'
)
```

# Versão POO

```python
class GradeProcessor:
    """Classe base para processamento de notas"""

    def __init__(self):
        pd.set_option('display.max_columns', None)

    def _load_student_data(self, student_file_path):
        """Carrega e padroniza os dados dos alunos"""
        df = pd.read_csv(student_file_path)
        return df.rename(columns={'NOME': 'Nome'})

    def _load_exercise_lists(self, list_files_prefix):
        """Carrega as listas de exercícios"""
        lists = []
        for i in range(1, 5):
            list_file = f"{list_files_prefix}{i}.csv"
            df_list = pd.read_csv(list_file)
            lists.append(df_list['Nome'].tolist())
        return lists

    def _update_list_status(self, df, lists):
        """Atualiza o status de entrega das listas"""
        for i in range(1, 5):
            df[f"l{i}"] = 0

        for i, exercise_list in enumerate(lists, 1):
            df[f"l{i}"] = df['Nome'].apply(lambda x: 1 if x in exercise_list else 0)
        return df

    def _calculate_grades(self, df, assessment_name):
        """Calcula as notas finais"""
        df[assessment_name] = df[['l1', 'l2', 'l3', 'l4']].sum(axis=1)
        df[f"n{assessment_name[-1]}"] = df[assessment_name] * 2.5
        return df

    def process_assessment(self, student_file_path, list_files_prefix, assessment_name, output_file):
        """Processa uma avaliação completa"""
        df = self._load_student_data(student_file_path)
        lists = self._load_exercise_lists(list_files_prefix)
        df = self._update_list_status(df, lists)
        df = self._calculate_grades(df, assessment_name)
        df = df.fillna(0)
        df.to_excel(output_file, index=False)
        print(f"Processamento concluído. Arquivo salvo em: {output_file}")


class AssessmentManager:
    """Gerencia múltiplas avaliações"""

    def __init__(self):
        self.processor = GradeProcessor()

    def run_assessments(self, assessments):
        """Executa o processamento para múltiplas avaliações"""
        for assessment in assessments:
            self.processor.process_assessment(
                student_file_path=assessment['student_file'],
                list_files_prefix=assessment['lists_prefix'],
                assessment_name=assessment['name'],
                output_file=assessment['output']
            )


# Configuração e execução
if __name__ == "__main__":
    assessments_config = [
        {
            'student_file': 'alunos_pe_20251.csv',
            'lists_prefix': 'av1/lista',
            'name': 'av1',
            'output': 'av1.xlsx'
        },
        {
            'student_file': 'alunos_pe_20251.csv',
            'lists_prefix': 'av2/lista',
            'name': 'av2',
            'output': 'av2.xlsx'
        }
    ]

    manager = AssessmentManager()
    manager.run_assessments(assessments_config)
```
