# O programa principal usando o Youtube-dl original

```python
import youtube_dl
import requests
```

```python
def download_subtitles(video_url):
    # Create a YouTubeDL instance
    ydl_opts = {'format': 'bestvideo[subtitles]/best[subtitles]', '--verbose': True}

    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        # Get video information
        video_info = ydl.extract_info(video_url, download=False)

        # Check if subtitles are available
        if 'subtitles' in video_info['formats'][0]:
            # Extract subtitles information
            subtitles = video_info['formats'][0]['subtitles']

            # Download the best available subtitle format
            for subtitle_format in subtitles.values():
                if subtitle_format['ext'] == 'srt':
                    subtitle_url = subtitle_format['url']
                    download_subtitle(subtitle_url, video_info['title'])
                    break

            # If no SRT subtitles found, display a message
            else:
                print(f"No SRT subtitles found for video: {video_info['title']}")
        else:
            print(f"Subtitles not available for video: {video_info['title']}")

def download_subtitle(subtitle_url, video_title):
    # Download the subtitle file
    with open(f"{video_title}.srt", 'wb') as f:
        response = requests.get(subtitle_url)
        f.write(response.content)

    print(f"Subtitle downloaded: {video_title}.srt")

def main():
    # Get video URL from the user
    video_url = input("Enter YouTube video URL: ")

    # Download subtitles using the download function
    download_subtitles(video_url)
```

```python
main()
```
