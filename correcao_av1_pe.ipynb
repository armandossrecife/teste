# Carrega arquivos

```python
def carrega_alunos(arquivo):
  l1 = list()
  try:
    f_l1 = open(arquivo, 'r')
    for linha in f_l1:
      l_temp = linha.split(' ')
      id = l_temp[0]
      matricula = l_temp[1]
      nome = l_temp[2:]
      temp = ' '.join(nome)
      temp = temp.strip()
      elemento = (id, matricula, temp)
      l1.append(elemento)
    f_l1.close() 
  except Exception as ex: 
    print(f'Erro ao carregar arquivo {arquivo}: {str(ex)}')
  return l1

def carrega_alunos_lista(arquivo):
  l1 = list()
  try:
    f_l1 = open(arquivo, 'r')
    for linha in f_l1:
      l1.append(linha.strip())
    f_l1.close() 
  except Exception as ex: 
    print(f'Erro ao carregar arquivo {arquivo}: {str(ex)}')
  return l1
```

```python
# lista com id, matricula e nome
l_pe = carrega_alunos('alunospe.txt') 
print(l_pe)
```

```python
# Carrega apenas os nomes dos alunos
f_alunos_pe = open('alunos_nomes_pe.txt', 'w')
for aluno in l_pe:
  temp = aluno[2] + '\n'
  f_alunos_pe.write(temp)
f_alunos_pe.close()
```

```python
# Carrega os nomes dos alunos que fizeram cada lista
lista1 = carrega_alunos_lista('l1.txt')
lista2 = carrega_alunos_lista('l2.txt')
lista3 = carrega_alunos_lista('l3.txt')
lista4 = carrega_alunos_lista('l4.txt')

print(f'Lista 1: {len(lista1)}, {lista1}')
print(f'Lista 2: {len(lista2)}, {lista2}')
print(f'Lista 3: {len(lista3)}, {lista3}')
print(f'Lista 4: {len(lista4)}, {lista4}')
```

# Faz a an√°lise...

```python
l_alunos_pe = []
f_alunos_pe = open('alunos_nomes_pe.txt')
for linha in f_alunos_pe:
  l_alunos_pe.append(linha.strip())
f_alunos_pe.close()

print(f'{len(l_alunos_pe)}, {l_alunos_pe}')
```

```python
f_alunos_csv = open('alunos_pe.csv', 'w')
f_alunos_csv.write('id, matricula, nome')
f_alunos_csv.write('\n')

for aluno in l_pe:
  id = aluno[0]
  matricula = aluno[1]
  nome_completo = aluno[2]
  nova_linha = id + ',' + matricula + ',' + nome_completo + '\n'
  f_alunos_csv.write(nova_linha)

f_alunos_csv.close()
```

```python
l_nome_completo = l_alunos_pe.copy()
print(f'{len(l_nome_completo)}, {l_nome_completo}')
```

```python
fez_l1, fez_l2, fez_l3, fez_l4 = ([], [], [], [])
for nome in l_nome_completo:
  if nome in lista1:  
    fez_l1.append(1)
  else:
    fez_l1.append(0)
  if nome in lista2:  
    fez_l2.append(1)
  else:
    fez_l2.append(0)
  if nome in lista3:  
    fez_l3.append(1)
  else:
    fez_l3.append(0)
  if nome in lista4:  
    fez_l4.append(1)
  else:
    fez_l4.append(0)
```

```python
dic_alunos_pe = {
    'nome':l_nome_completo,
    'fez_l1':fez_l1, 
    'fez_l2':fez_l2, 
    'fez_l3':fez_l3, 
    'fez_l4':fez_l4, 
}
```
