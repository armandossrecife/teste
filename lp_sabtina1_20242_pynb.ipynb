```python
def limpa_nomes_alunos(lista_nomes_alunos):
  lista_nomes_alunos_limpos = []
  for each in lista_nomes_alunos:
    each = each.replace("\t", "")
    lista_nomes_alunos_limpos.append(each)

  return lista_nomes_alunos_limpos

def gera_lista_respostas(arquivo):
  with open(arquivo) as f_lista:
    lista_alunos = f_lista.read().splitlines()
  return lista_alunos

def gera_dicionario_alunos_lista(dicionario_alunos, lista_alunos_lista):
  dicionario_alunos_lista = {}

  for matricula, nome in dicionario_alunos.items():
    if nome in lista_alunos_lista:
      dicionario_alunos_lista[matricula] = nome

  return dicionario_alunos_lista

def checa_alunos_fezeram_lista(df_alunos, lista_alunos_lista):
  lista1_respostas = []
  for each in lista_alunos_lista:
    if each in df_alunos.nome.tolist():
      print(f"Aluno {each} fez a lista")
      lista1_respostas.append(1)
    else:
      print(f"Aluno {each} não fez a lista")
      lista1_respostas.append(0)
```
