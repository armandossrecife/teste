# Dados do SIGAA

```python
import pandas
```

```python
df_lista1 = pandas.read_csv('respoastal1.csv', sep=';')
df_lista2 = pandas.read_csv('respoastal2.csv', sep=';')
df_lista3 = pandas.read_csv('respoastal3.csv', sep=';')
df_lista4 = pandas.read_csv('respoastal4.csv', sep=';')
df_alunos = pandas.read_csv('alunostes20242.csv', sep=';')
```

```python
df_alunos.head(3)
```

```python
lista1 = df_lista1['Aluno'].tolist()
lista2 = df_lista2['Aluno'].tolist()
lista3 = df_lista3['Aluno'].tolist()
lista4 = df_lista4['Aluno'].tolist()
```

```python
# prompt: how to convert df_alunos do dictionary where key is Matricula and value of item is Nome?

dicionario_alunos = df_alunos.set_index('Matricula')['Nome'].to_dict()
```

```python
dicionario_alunos
```

```python
dl1, dl2, dl3, dl4 = dict(), dict(), dict(), dict()

for k, v in dicionario_alunos.items():
    if v in lista1:
        print(f'Aluno {v} está na lista 1')
        dl1[k] = 1
    else:
        dl1[k] = 0
    if v in lista2:
        print(f'Aluno {v} está na lista 2')
        dl2[k] = 1
    else:
        dl2[k] = 0
    if v in lista3:
        print(f'Aluno {v} está na lista 3')
        dl3[k] = 1
    else:
        dl3[k] = 0
    if v in lista4:
        print(f'Aluno {v} está na lista 4')
        dl4[k] = 1
    else:
        dl4[k] = 0
```

```python
# prompt: how to iterate df_alunos using for?
valor_l1, valor_l2, valor_l3, valor_l4 = [], [], [], []

for index, row in df_alunos.iterrows():
  # Access individual row elements using column names
  matricula = row['Matricula']
  nome = row['Nome']
  # Perform operations on the row data
  print(f"Matricula: {matricula}, Nome: {nome}")
  if matricula in dl1.keys():
    valor_l1.append(dl1[matricula])
  if matricula in dl2.keys():
    valor_l2.append(dl2[matricula])
  if matricula in dl1.keys():
    valor_l3.append(dl3[matricula])
  if matricula in dl1.keys():
    valor_l4.append(dl4[matricula])
```

```python
d_alunos_tes_20242 = {
    'Matricula': df_alunos['Matricula'],
    'Nome': df_alunos['Nome'],
    'Lista1': valor_l1,
    'Lista2': valor_l2,
    'Lista3': valor_l3,
    'Lista4': valor_l4,
}
```

```python
df_alunos_tes_20242 = pandas.DataFrame(d_alunos_tes_20242)
df_alunos_tes_20242
```

```python
# prompt: In dataframe df_alunos_tes_20242, how to create a new column to record the sum of columns lista1, lista2, lista3 and lista4?

df_alunos_tes_20242['Soma_Listas'] = df_alunos_tes_20242['Lista1'] + df_alunos_tes_20242['Lista2'] + df_alunos_tes_20242['Lista3'] + df_alunos_tes_20242['Lista4']
```

```python
df_alunos_tes_20242
```

```python
df_alunos_tes_20242['N1'] = df_alunos_tes_20242['Soma_Listas']*2
```

```python
# prompt: In dataframe df_respostas_aws_t, except columns Matricula and Student, if the value of column is different of zero you have to put number 1?

for column in df_respostas_aws_t.columns:
  if column not in ['Matricula', 'Student']:
    df_respostas_aws_t[column] = df_respostas_aws_t[column].apply(lambda x: 1 if x != 0 else 0)
```

```python
# prompt: In dataframe df_respostas_aws_t, except columns Matricula and Student, sum all values of the columns.

df_respostas_aws_t['Soma_Atividades'] = df_respostas_aws_t.iloc[:, 2:].sum(axis=1)
```

```python
colunas_uteis_aws = ['Matricula', 'Student', 'Soma_Atividades']
df_atividades = df_respostas_aws_t[colunas_uteis_aws]
df_atividades.sort_values(by=['Soma_Atividades'], ascending=False)
```

```python
# prompt: Para o dataframe df_atividades, Na coluna Soma_Atividades, crie uma função que dada as seguintes faixas de valores substitua pelos valores de cada faixa
# faixa1 = 0 substitua por 0
# faixa2 = 1 a 3 substitua por 0.5
# faixa3 = 4 a 5 substitua por 1
# faixa4 = 6 a 10 substitua por 1.4
# faixa5 = 11 a 12 substitua por 1.6
# faixa6 = 13 substitua por 1.8
# faixa7 = 14 substitua por 2

def faixa_atividades(soma_atividades):
  if soma_atividades == 0:
    return 0
  elif 1 <= soma_atividades <= 3:
    return 0.5
  elif 4 <= soma_atividades <= 5:
    return 1
  elif 6 <= soma_atividades <= 10:
    return 1.4
  elif 11 <= soma_atividades <= 12:
    return 1.6
  elif soma_atividades == 13:
    return 1.8
  elif soma_atividades == 14:
    return 2
  else:
    return None  # Ou outro valor padrão caso necessário

df_atividades['Nota_Atividades'] = df_atividades['Soma_Atividades'].apply(faixa_atividades)
```
