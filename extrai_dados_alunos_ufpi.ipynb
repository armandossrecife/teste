```python
import requests
```

```python
URL_ALUNOS_COMPUTACAO_UFPI = "https://www.sigaa.ufpi.br/sigaa/public/curso/alunos.jsf?lc=pt_BR&id=74268"

alunos_computacao_ufpi = requests.get(URL_ALUNOS_COMPUTACAO_UFPI)
```

```python
if alunos_computacao_ufpi.status_code:
    print("Requisição bem sucedida")
    conteudo = alunos_computacao_ufpi.content
    print(conteudo)
else:
    print("Requisição falhou")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(conteudo, 'html.parser')
table = soup.find('table')

if table:
  # Extrair dados da tabela
  rows = table.find_all('tr')
  for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    print(cols)
else:
  print("Tabela não encontrada na página.")
```
