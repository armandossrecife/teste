```python
import networkx as nx

def criar_grafo(nos, arestas, direcionado):
  if not direcionado:
    # Criar um grafo não direcionado
    G = nx.Graph()
  else:
    G = nx.DiGraph()

  # Adicionar os nós
  G.add_nodes_from(nos)

  # Adicionar as arestas (amizades)
  G.add_edges_from(arestas)

  # Visualizar o grafo
  nx.draw(G, with_labels=True)
```

```python
lista_nos = ['A', 'B', 'C', 'D', 'E']
lista_arestas = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E')]

criar_grafo(nos=lista_nos, arestas=lista_arestas, direcionado=False)
```

```python
criar_grafo(nos=lista_nos, arestas=lista_arestas, direcionado=True)
```
